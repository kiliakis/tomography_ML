```python
import pickle as pk
import pandas as pd
import os
import glob
```

```python
hparam_trials_list = [
    './hparam_trials/2023_03_28*',
    './hparam_trials/2023_04_19*',
    './hparam_trials/2023_04_20*',
    ]
```

```python
# expand regexps in list
temp_list = []
for trial in hparam_trials_list:
    temp_list += glob.glob(trial)
hparam_trials_list = temp_list

for trial in hparam_trials_list:
    print(trial)
```

```python
all_temp_data = []
for triali, trial in enumerate(hparam_trials_list):
    dir = os.path.join(trial, 'hparams')

    for file in os.listdir(dir):
        variable = file.split('_')[0]
        path = os.path.join(dir, file)
        data = pk.load(open(path, 'rb'))
        all_data_d = {}
        for k, v in data.items():
            unique_key = f'{variable}-{triali}-{k}'
            if 'run' not in all_data_d:
                all_data_d['run'] = []
            all_data_d['run'].append(k.split('-')[1])
            # if 'run' not in all_data_d:
            #     all_data_d['run'] = []
            # all_data_d['run'].append(k.split('-')[1])
            if 'var' not in all_data_d:
                all_data_d['var'] = []
            all_data_d['var'].append(variable)
            for cat, value in v.items():
                if cat == 'history':
                    epochs = len(v['history']['loss'])
                    if 'epochs' not in all_data_d:
                        all_data_d['epochs'] = []
                    all_data_d['epochs'].append(epochs)
                else:
                    if isinstance(value, list):
                        value = '-'.join(str(e) for e in value)
                    if cat not in all_data_d:
                        all_data_d[cat] = []
                    all_data_d[cat].append(value)
        temp_data = pd.DataFrame(all_data_d)
        all_temp_data.append(temp_data)
data = pd.concat(all_temp_data)
data.head()
data = data.sort_values(by=['var', 'valid'])
```
